## Caminata Aleatoria Simple
---

Se genera una camina aleatoria simple, donde el rango de movimiento sobre la línea de los números ℤ es de $1$ y $-1$, con probilidades $p$ y $q$, las cuáles $p+q=1$.

```python
#Librerías a utilzar
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Función del paseo aleatorio
def randomwalk1D(n):
    x, y = 0, 0
    # Generamos los puntos [1, 2, 3, ... , n]
    timepoints = np.arange(n + 1)
    positions = [y]
    directions = ["UP", "DOWN"]
    for i in range(1, n + 1):
        # Seleccionamos al azar arriba o abajo
        step = random.choice(directions)

        # Movemos el objeto arriba o abajo
        if step == "UP":
            y += 1
        elif step == "DOWN":
            y -= 1
        # Registramos la posición del objeto
        positions.append(y)
    return timepoints, positions
```

```python
#Número de pasos
n_steps = 10000
time_data, pos_data = randomwalk1D(n_steps)
#Gráfica de camina aleatoria simple
plt.plot(time_data, pos_data, 'r-')
plt.title("Caminata Aleatoria 1-D")
plt.xlabel("Tiempo")
plt.ylabel("Pasos")
plt.grid(True)
plt.show()
```

### Análisis y Conclusiones

Como podemos ver al ser un proceso estocástico, dado que contamos un parámtro $t$, donde se mueve el objeto, tendremos que con sólo dos pasos, ya sea de derecha con probilidad $p$ y a la izquierda con probabilidad $q$, siendo que entonces observamos que la posición del objeto es la suma en base a los pasos anteriores para saber la posición final.
